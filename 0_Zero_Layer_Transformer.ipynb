```python
import torch
import torch.nn as nn

class ZeroLayerTransformer(nn.Module):
    def __init__(self, vocab_size, d_model):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, d_model)
        self.unembed = nn.Linear(d_model, vocab_size, bias=False)

    def forward(self, tokens):
        x = self.embed(tokens)      # tokens → vectors # x = W^E token
        logits = self.unembed(x)    # vectors → vocab scores # W^U W^E token
        return logits
```

```python
vocab_size = 1000
d_model = 64

model = ZeroLayerTransformer(vocab_size, d_model)
tokens = torch.tensor([[1, 5, 3, 7, 2]])  # fake token ids
logits = model(tokens)
print(f"Input shape:  {tokens.shape}")
print(f"Output shape: {logits.shape}")
print(f"Prediction at last position: {logits[0, -1, :].argmax()}")
```

```python
embedded_tokens = model.embed(tokens)
print(f"Embedded tensor shape: {embedded_tokens.shape}")
print(f"Embedded tensor content (first 5 elements of first token):\n{embedded_tokens[0, 0, :5]}")
```

```python
raw_embeddings = model.embed.weight
print(f"Shape of raw embedding weights: {raw_embeddings.shape}")
print(f"First 5 elements of the embedding vector for token ID 0:\n{raw_embeddings[0, :5]}")
print(f"First 5 elements of the embedding vector for token ID 1:\n{raw_embeddings[1, :5]}")
```

This `raw_embeddings` tensor has a shape of `(vocab_size, d_model)`, where each row represents the embedding vector for a specific token ID. For example, `raw_embeddings[0]` is the embedding vector for token ID 0, and `raw_embeddings[1]` is the embedding vector for token ID 1, and so on.
